```python
report_card = {"Student_names":["Vanshika","Vivaan","Siddhant","Mitansh","Kareena"],
               "physics_grades":[38,45,65,75,80],
               "chemistry_grades":[100,45,56,67,43],
               "maths_grades":[90,80,70,60,50],
               "biology_grades":[10,20,30,40,50]
              }
report_card
```

```python
#setup
import pandas as pd
```

```python
#making the report card dictionary into dataframe
report_card_df = pd.DataFrame(report_card)
report_card_df
```

# Insights
 - what is the average chemistry grade? 62.2
 - what is the average biology grade? 30,0
 - what is the is the lowest math grade? 50.0
 - what is the highest physics grade? 80.0
 - How many scores are there in the math coloumn? 5

```python
# Making a total grade coloumn
report_card_df["total_grade"] = report_card_df["physics_grades"]+ report_card_df["chemistry_grades"]+ report_card_df["maths_grades"]+ report_card_df["biology_grades"]
report_card_df
```

```python
report_card_df["average_grade"] = report_card_df["total_grade"]/4
report_card_df
```

```python
report_card_df.describe()
```

```python
report_card_insights_df = report_card_df.describe().transpose()
report_card_insights_df
```

```python
report_card_insights_df.to_csv("Report card.csv")
```

```python
report_card_df.loc[report_card_df["total_grade"]>=140, "Status"]="Pass"
report_card_df.loc[report_card_df["total_grade"]<140, "Status"]="Fail"
report_card_df
```

```python
report_card_insight_df = report_card_df
report_card_insight_df.to_csv("Report card.csv")
```

```python
report_card_df.loc[(report_card_df["average_grade"] >=90), "Character_grade"]="A"
report_card_df.loc[(report_card_df["average_grade"] >=70) & (report_card_df["average_grade"]<90),"Character_grade"]="B"
report_card_df.loc[(report_card_df["average_grade"] >=50) & (report_card_df["average_grade"]<69),"Character_grade"]="C"
report_card_df.loc[(report_card_df["average_grade"] >=40) & (report_card_df["average_grade"]<49),"Character_grade"]="D"
report_card_df.loc[(report_card_df["average_grade"] <40), "Character_grade"]= "F"
```
